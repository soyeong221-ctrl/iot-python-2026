#### 위경도 특성
- 소수점 4자리이상 작성해야 제대로된 위치를 잡을 수 있음

````python
# folium 라이브러리 로드
import folium

# 37.551192, 126.988165(N 서울타워)
````

````python
loc = [35.11745, 129.0907]

# 지도 객체 생성, m 변수 할당
m = folium.Map(location=[35.11745, 129.0907], zoom_start=15)

# 마커를 생성해서 지도변수에 추가
folium.Marker(location=loc,
            tooltip='부산',
            popup='용당7관',
            icon=folium.Icon(icon='home', prefix='fa')).add_to(m)

m
````
