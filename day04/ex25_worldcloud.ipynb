#### 기본 사용

````python
# wordcloud 라이브러리 로드
from wordcloud import WordCloud
import matplotlib.pyplot as plt
````

````python
text = ''' python is my favorite programming language, data analysis, python ai ai ai
            deep learning, machine learning, ai contemporary ai skill
        '''

wc = WordCloud().generate(text)
````

````python
plt.imshow(wc)
plt.axis('off') # 차트 x, y축 눈금, 레이블 의미하는 객체
plt.show()
````

````python
# C++, C#의 특수문자를 제대로 인식 못함
text = '데이터 분석 파이썬 인공지능 인공지능 인공지능 데이터 파이썬 C++ Java 인공지능 C#'

wc = WordCloud(font_path='C:/Windows/Fonts/NanumGothicCoding.ttf',
               background_color='azure',
               colormap='plasma'  # viridis, plasma, inferno, cool
               ).generate(text)
               
plt.imshow(wc)
plt.axis('off') # 차트 x, y축 눈금, 레이블 의미하는 객체
plt.show()
````
