````python
# OpenCV 라이브러리 로드, 넘파이가 필요할 수 있음
import cv2
import numpy as np
````

````python
# 이미지로드
img = cv2.imread('./cat01.png')

cv2.imshow('origin', img)   # origin 윈도우 제목으로 이미지 오픈
cv2.waitKey(0)              # 키보드 입력 대기, 무슨키든지 다 종료
cv2.destroyAllWindows()     # 종료 후 열린 모든 윈도우창 닫음
````

````python
# 동영상로드
cap = cv2.VideoCapture(filename='./sample.mp4')

while True: # 무한루프
    ret, frame = cap.read() # FPS의 이미지 하나씩 읽어옴
    if not ret: break # 더 이상 읽을 이미지가 없으면 종료

    cv2.imshow('video', frame) # 이미지를 계속 변경
    if cv2.waitKey(1) == ord('q'): break    # 키보드 q 입력하면 종료

cap.release()   # 메모리 해제    
cv2.destroyAllWindows()
````

````python
# 실시간 웹캠
cap = cv2.VideoCapture(0)   # 웹캠 디바이스 번호

while True: 
    ret, frame = cap.read() 
    if not ret: break

    cv2.imshow('webcam', frame) 
    if cv2.waitKey(1) == ord('q'): break    # 키보드 q 입력하면 종료

cap.release()   # 메모리 해제    
cv2.destroyAllWindows()
````

#### 이미지 처리

````python
# 300x300 크기 빨강이미지 만들기
# np.zeros((width pixel, height pixel, channel:R G B))
img = np.zeros((300, 300, 3), dtype=np.uint8)
img[:] = (0, 0, 255) # OpenCV RGB를 거꾸로 사용: B, G, R

cv2.imwrite('./red_img.jpg', img)
````

````python
# 이미지 특정 위치 변경
# imread 옵션 색상변경, 사이즈변경
cat = cv2.imread(filename='./cat01.png')

# 이미지 영역
# 고양이 이미지에서 특정 영역 지정
roi = cat[168:168 + 35, 318:318 + 192] # [행:행 + 높이, 열:열 + 넓이]
roi[:] = (0, 0, 0)  # 검은색 리슽, 튜플 사용가능

cv2.imshow('cat parasite', cat)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# imread 옵션
# flags
# IMREAD_COLOR, IMREAD_ANYCOLOR, IMREAD_COLOR_RGB = 색상 그대로
# IMREAD_COLOR_RGB = OpenCV 색상순서 거꾸로 읽기
# IMREAD_GRAYSCALE = 회색톤
# IMREAD_REDUCED_GRAYSCALE_2 = 회색톤 1/2 사이즈
# IMREAD_REDUCED_COLOR_2 = 컬러 1/2 사이즈
cat = cv2.imread(filename='./cat01.png', flags=cv2.IMREAD_COLOR)

# 이미지 영역
# 고양이 이미지에서 특정 영역 지정
roi = cat[168:168 + 35, 318:318 + 192] # [행:행 + 높이, 열:열 + 넓이]
roi[:] = (0, 0, 0)  # 검은색 / 0 GrayScale은 색이 하나

cv2.imshow('cat parasite', cat)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# 이미지 로드 후 변경
# 로드한 이미지는 매트릭스 객체에 할당
cat = cv2.imread(filename='./cat01.png', flags=cv2.IMREAD_COLOR)

# OpenCV 함수로 변경
gray = cv2.cvtColor(cat, cv2.COLOR_BGR2GRAY)    # 칼라를 회색톤으로 변경

cv2.imshow('origin', cat)
cv2.imshow('gray', gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# 채널 분리
cat = cv2.imread(filename='./cat01.png', flags=cv2.IMREAD_COLOR)

(b, g, r) = cv2.split(cat)

# 채널(B,G,R)를 단일 채널로 분리
# np.zeros(): 배열 사이즈를 직접 지정해서 0으로 배열초기화
# np.zeros_like(배열): 지정된 배열사이즈를 0으로 초기화
blue = cv2.merge([b, np.zeros_like(g), np.zeros_like(r)])
green = cv2.merge([np.zeros_like(b), g, np.zeros_like(r)])
red = cv2.merge([np.zeros_like(b), np.zeros_like(g), r])

cv2.imshow('origin', cat)
cv2.imshow('Blue', blue)
cv2.imshow('green', green)
cv2.imshow('red', red)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# 이미지 로드 후 사이즈 변경
cat = cv2.imread(filename='./cat01.png', flags=cv2.IMREAD_COLOR)

# OpenCV 함수로 변경
small = cv2.resize(cat, None, fx=0.5, fy=0.5)

cv2.imshow('origin', cat)
cv2.imshow('1/2 size', small)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# 이미지 로드 후 회전
cat = cv2.imread(filename='./cat01.png', flags=cv2.IMREAD_COLOR)

# OpenCV 함수로 변경
rot180 = cv2.rotate(cat, cv2.ROTATE_180)

cv2.imshow('origin', cat)
cv2.imshow('rot180', rot180)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

#### OpenCV 이미지 처리 주요기능

- 주요기능 - 포토샵 등 이미지편집기 기능과 동일
    - 이미지 읽기, 저장, 출력
    - 이미지 합성, 이미지 자르기
    - 색상변환
    - 크기변경
    - 회전, 이동
    - 블러
    - 이진화
    - 선명도 조절
    - 노이즈 제거
    - 대비/밝기 조절
    - 엣지 검출
    - 윤곽선 찾기
    - 퍼시펙티브 변환 - 원근감 이미지를 평탄화

````python
# 이미지 로드
cat = cv2.imread(filename='./cat01.png', flags=cv2.IMREAD_COLOR)

# 원 그리기
# OpenCV 이미지 자체는 Mat(매트릭스)라서 행(height), 열(width) 순으로 처리
# 원 포인트는 (x, y)로 처리
# thickness=-1 원 내부 채우기
cv2.circle(cat, (416, 210), 100, (255, 255, 255), thickness=5, lineType=cv2.LINE_AA)

cv2.imshow('origin', cat)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# 이미지 로드
cat = cv2.imread(filename='./cat01.png', flags=cv2.IMREAD_COLOR)

# 텍스트
cv2.putText(cat, 'Kitty', (300, 70),
            cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 255, 255), 3, cv2.LINE_AA)
cv2.putText(cat, '♥', (300, 70),
            cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 255), 3, cv2.LINE_AA)

cv2.imshow('origin', cat)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

### ROI

- Region Of Interest: 관심영역
    - 2차원리스트(매트릭스, 행렬)
    - 영역추출법 - Mat[y시작점:y시작점+높이, x시작점:x시작점+넓이]

````python
# 이미지 로드
cat = cv2.imread(filename='./cat01.png', flags=cv2.IMREAD_COLOR)

# ROI 선택
roi = cat[81:81+214, 295:295+240]

cv2.imshow('origin', cat)
cv2.imshow('roi', roi)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# 이미지 로드
cat = cv2.imread(filename='./cat01.png', flags=cv2.IMREAD_COLOR)

## 엣지(이미지 경계) 검출
# 그레이스케일 변경
gray = cv2.cvtColor(cat, cv2.COLOR_BGR2GRAY)

# canny 엣지
# threshold 임계값, 1은 약한 엣지, 2는 강한 엣지
canny = cv2.Canny(gray, threshold1=250, threshold2=300)

cv2.imshow('gray', gray)
cv2.imshow('canny', canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
````
